```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass


    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass


    return False
```

```python
def map_dict(filename):
    """
    get srt file and return a dictionary with 
    start-time_end-time keys and 
    phrases for values
    """
    with open(filename) as f:
        content = f.readlines()
    content = [x.strip() for x in content]
    l=len(content)
    times={}
    for i in range(l - 2):
        try:
            if is_number(content[i]): #finds counter line
                a=content[i+1].split('-->')
                key=a[0].split(',')[0]+'_'+a[1].split(',')[0] #get start time & end time
                c=i+2
                while is_number(content[c])!=True: 
                    if(c<l):
                        if key in times:
                            times[key] += content[c]
                            if(c<l - 1):
                                c+=1
                            else:
                                break
                        else:
                            times[key] = content[c]
                            if(c<l - 1):
                                c+=1 
                            else:
                                break
        except:
                print("list index erorr")
    return times
           

```
